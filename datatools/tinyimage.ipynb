### Reading, resizing and saving images with `scipy`
One way to create some tinyimages ([MIT Project link](http://groups.csail.mit.edu/vision/TinyImages/)) in Python from pratically any image.  Donating to data science.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc

%matplotlib inline
```

**Showing a stock image with `matplotlib`**

```python
face = misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
```

**Resizing and saving that image, plus another**

```python
face_small = misc.imresize(face, (32,32))
plt.imshow(face_small)
```

```python
misc.imsave('face.png', face_small)
```

```python
car = misc.imread('metal-model-toy-car.jpg')
plt.imshow(car)
car_small = misc.imresize(car, (32,32))
plt.imshow(car_small)
```

```python
misc.imsave('car_small.png', car_small)
```
